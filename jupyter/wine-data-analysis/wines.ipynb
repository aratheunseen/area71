## Load data

```python
import pandas as pd
```

```python
df = pd.read_csv("wines.csv")
```

```python
df
```

## Show total wines with 100 points rating

```python
len(df[df['points']==100])
```

## Show most expensive wine name

```python
df.loc[df['price'].max() == df['price']]['name'].squeeze()
```

## Convert points to 0-5 rating system

```python
df['new_rating'] = df['points'] / 100 * 5
df
```

## Show hostogram of wines that cost 100 or less

```python
df[df['price'] <= 100]['price'].hist()
```

## Plot price horizontally against points vertically

```python
df.plot(x='price',y='points',kind="line")
df.plot(x='price',y='points',kind="scatter")
```
